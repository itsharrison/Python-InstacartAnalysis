```python
# Import ords_prods_merge
df_ords_prods_merge = pd.read_pickle(os.path.join(path, 'Data', 'Prepared Data', 'ords_prods_merge.pkl'))
```

```python
# Create a subset of the dataframe
df = df_ords_prods_merge[:1000000]
```

```python
df
```

```python
# Set up a function called price_label
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check the most expensive product
df['prices'].max()
```

```python
# Set up a loc function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Run a loc function on the entire dataframe
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_ords_prods_merge['price_range_loc'].value_counts(dropna = False)
```

```python
# Find out which day most orders take place
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

Most orders take place on a Saturday.

```python
# Create an empty list that can contain different results
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create a new column with the new information
df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
# 0 and 1 are the busiest days. 3 and 4 are the slowest days. Create another empty list that can contain the new required results
result_2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 3 or value == 4:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
result_2
```

```python
# Create a new column with the new information
df_ords_prods_merge['busiest_days'] = result_2
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# Find out wchich hour most orders take place
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

The hours should be divided into three periods equally, which are "Most orders", "Average orders", and "Fewest orders".

```python
# Create a for loop for different periods
result_3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
    result_3.append("Average orders")
  else:
    result_3.append("Fewest orders")
```

```python
result_3
```

```python
# Create a new column with the new information
df_ords_prods_merge['busiest_period_of_day'] = result_3
```

```python
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
