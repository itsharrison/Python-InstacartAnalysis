## 2. Visualizations

```python
# Create a bar chart
ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
plt.title('Frequency of Orders on all Days of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Order Frequency')
```

```python
# Sort the bar chart according to the index order
bar_frequency_dow = ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
plt.title('Frequency of Orders on all Days of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Order Frequency')
```

```python
# Save the bar chart within the bar object
bar_frequency_dow.figure.savefig(os.path.join(path, 'Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
# Save the bar chart within the bar object
bar_colored_dow = ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
plt.title('Frequency of Orders on all Days of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Order Frequency')
```

```python
# Save the bar chart in the visualizations folder
bar_colored_dow.figure.savefig(os.path.join(path, 'Analysis','Visualizations', 'bar_colored_orders_dow.png'))
```

```python
# Create a histogram of the prices column again 
hist = ords_prods_merge['prices'].plot.hist(bins = 25)
plt.title('Price Histogram')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
# Increase the granularity
hist = ords_prods_merge['prices'].plot.hist(bins = 70)
plt.title('Price Histogram')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

```python
# Save the histogram in the visualizations folder
hist.figure.savefig(os.path.join(path, 'Analysis','Visualizations', 'hist_prices.png'))
```
